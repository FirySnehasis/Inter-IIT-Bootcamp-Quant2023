```python
!pip install vectorbt
import pandas as pd
import numpy as np
import yfinance as yf
import math
import matplotlib.pyplot as plt
import vectorbt as vbt
```

```python
data = yf.download('AAPL', start='2016-01-01', end='2023-06-15')
```

```python
data
```

```python
data['Buy_Signal'] = 0
data['Sell_Signal'] = 0
data['MACD'] = data['Close'].ewm(span=12, adjust=False).mean() - data['Close'].ewm(span=26, adjust=False).mean()
data['Signal Line MACD'] = data['MACD'].ewm(span=9, adjust=False).mean()
data['50SMA'] = data['Close'].ewm(span=95, adjust=False).mean()
for i in range(1,len(data)-1):
  if((data.iloc[i,9]>=data.iloc[i,8] and data.iloc[i+1,8]>=data.iloc[i,9])):
    data.iloc[i, 6] = 1

  elif(data.iloc[i,9]<=data.iloc[i,8] and data.iloc[i+1,8]<=data.iloc[i,9]):
    data.iloc[i, 7] = -1

portfolio1 = vbt.Portfolio.from_signals(data['Close'], data['Buy_Signal'], data['Sell_Signal'], init_cash=10000000, fees=0.001, freq='D')
print(portfolio1.stats())
portfolio1.plot().show()
portfolio1.plot_trades().show()
```
