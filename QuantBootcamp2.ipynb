# Here we are selling and buying the stocks with the help of MACD indiacator and then we are backtesting it with the help of *VectorBT* .

```python
data['Buy_Signal'] = 0
data['Sell_Signal'] = 0
for i in range(1,len(data)-1):
  if(data1.loc[i,'signal_line']>=data1.loc[i,'macd_line'] and data1.loc[i+1,'macd_line']>=data1.loc[i,'signal_line']):
    data.iloc[i, 20] = 1

  elif(data1.loc[i,'signal_line']<=data1.loc[i,'macd_line'] and data1.loc[i+1,'macd_line']<=data1.loc[i,'signal_line']):
    data.iloc[i, 21] = -1

portfolio1 = vbt.Portfolio.from_signals(data['close'], data['Buy_Signal'], data['Sell_Signal'], init_cash=10000000, fees=0.001, freq='D')
print(portfolio1.stats())
portfolio1.plot().show()
portfolio1.plot_trades().show()
```

# Here we are selling and buying the stocks with the help of Bollinger Bands and then we are backtesting it with the help of *VectorBT* .

```python
data['Buy_Signal2'] = 0
data['Sell_Signal2'] = 0
for i in range(1,len(data)-1):
  if(abs((data.iloc[i,6]-data2.iloc[i,3])/(data2.iloc[i,3]))<0.05):
    data.iloc[i,23] = -1
  elif(abs((data.iloc[i,6]-data2.iloc[i,1])/(data2.iloc[i,1]))<0.05):
    data.iloc[i,22] = 1

portfolio2 = vbt.Portfolio.from_signals(data['close'], data['Buy_Signal2'], data['Sell_Signal2'], init_cash=10000000, fees=0.001, freq='D')
print(portfolio2.stats())
portfolio2.plot().show()
portfolio2.plot_trades().show()
```
